```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data20= pd.read_csv('gastos_costos_20.csv')
```

```python
#Conocer estructura y eliminar columnas si es necesario
data20 = data20.drop(['Unnamed: 0','Columna1'], axis=1)
data20.info()
```

```python
#Filtrar por cuantitativas (Atipicos)
cuantitativas=data20.iloc[:,8:14]
cuantitativas
```

```python
#Filtrar por cualitativas
cualitativas=data20.iloc[:,[0,1,2,3,4,5,6,7,15,16,17]]
cualitativas
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data20.isnull().sum()
valores_nulos
```

```python
#MÃ©todo aplicando desviaciÃ³n estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2020= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2020
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2020.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atÃ­picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data2020.copy()
data_clean=data_clean.fillna(round(data2020.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe original
fig = plt.figure(figsize =(15, 8))
data20.plot(kind='box', vert=False)
plt.title("Valores AtÃ­picos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe limpio
fig = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores AtÃ­picos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Unimos las columnas cualitativas con el dataframe
Datos_limpios = pd.concat([cualitativas, data_clean], axis=1)
Datos_limpios
```

```python
#Convertir archivo filtrado a CSV
Datos_limpios.to_csv("Valores AtÃ­picos (Gastos y costos 2020).csv")
```

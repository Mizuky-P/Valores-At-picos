```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data22= pd.read_csv('gastos_costos_22.csv')
```

```python
#Conocer estructura y eliminar columnas si es necesario
data22 = data22.drop(['Unnamed: 0'], axis=1)
data22.info()
```

```python
#Filtrar por cuantitativas (Atipicos)
cuantitativas = data22.iloc[:, [8, 9, 10, 11, 12, 14, 15]]
cuantitativas
```

```python
#Filtrar por cualitativas
cualitativas=data22.iloc[:,[0,1,2,3,4,5,6,7,13,16,17,18]]
cualitativas
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data22.isnull().sum()
valores_nulos
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data2022= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2022
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2022.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data2022.copy()
data_clean=data_clean.fillna(round(data2022.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe original
fig = plt.figure(figsize =(15, 8))
data22.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe limpio
fig = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Unimos las columnas cualitativas con el dataframe
Datos_limpios = pd.concat([cualitativas, data_clean], axis=1)
Datos_limpios
```

```python
#Convertir archivo filtrado a CSV
Datos_limpios.to_csv("Valores Atípicos (Gastos y costos 2022).csv")
```

```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
```

```python
#Carga desde un archivo .csv sin indice
data23= pd.read_csv('gastos_costos_23.csv')
```

```python
#Conocer estructura y eliminar columnas si es necesario
data23 = data23.drop(['Unnamed: 0'], axis=1)
data23.info()
```

```python
#Filtrar por cuantitativas (Atipicos)
cuantitativas = data23.iloc[:,8:16]
cuantitativas
```

```python
#Filtrar por cualitativas
cualitativas=data23.iloc[:,[0,1,2,3,4,5,6,7,17,18,19]]
cualitativas
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data23.isnull().sum()
valores_nulos
```

```python
#M√©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data2023= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2023
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2023.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores at√≠picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data2023.copy()
data_clean=data_clean.fillna(round(data2023.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Unimos las columnas cualitativas con el dataframe
Datos_limpios = pd.concat([cualitativas, data_clean], axis=1)
Datos_limpios
```

```python
#Convertir archivo filtrado a CSV
Datos_limpios.to_csv("Valores At√≠picos (Gastos y costos 2023).csv")
```
